```python
from datetime import timedelta,datetime

class Hotel:
    
    def __init__(self,hotelName,hotelAddress):
        self._hotelName = hotelName
        self._hotelAddress = hotelAddress
    
    @property
    def hotelName(self):
        return self._hotelName
    @property
    def hotelAddress(self):
        return self._hotelAddress
    
    
    def __str__(self):
        return f"Hotel Name: {self._hotelName:<7} Address: {self._hotelAddress:<6}"

class Driver:
    
    def __init__(self,driverName,vehicleNum):
        
        self._driverName = driverName
        self._vehicleNum = vehicleNum
    
    @property
    def driverName(self):
        return self._driverName
    
    @property
    def vehicleNum(self):
        return self._vehicleNum
    @vehicleNum.setter
    def vehicleNum(self,newValue):
        self._vehicleNum = newValue
    
    def __str__(self):
        return f"Driver Name: {self._driverName:<7} Vehicle Number: {self._vehicleNum:<6}"
    
class Booking:
    
    def __init__(self,datetimeOfPickup,fromAirport,numOfpassangers,hotel):
        
        self._datetimeOfPickup = datetimeOfPickup
        self._fromAirport = fromAirport
        self._numofPassangers = numOfpassangers
        self._hotel = hotel
        self._driver = None
        
    @property
    def datetimeOfPickup(self):
        return self._datetimeOfPickup
    @datetimeOfPickup.setter
    def datetimeOfPickup(self,newValue):
        self._datetimeOfPickup = newValue
    @property
    def fromAirport(self):
        return self._fromAirport
    @property
    def numOfpassangers(self):
        return self._numofPassangers
    @numOfpassangers.setter
    def numOfpassangers(self,newValue):
        self._numofPassangers = newValue
    @property
    def hotel(self):
        return self._hotel
    @property
    def driver(self):
        return self._driver
    @driver.setter
    def driver(self,newValue):
        self._driver = newValue
    
    def __str__(self):
        
        if self._fromAirport:
            way = 'To Hotel'
        else:
            way = 'To Airport'
        if not self._driver:
            driver = 'NO'
        else:
            driver = self._driver
            
        return f"Date/Time: {self._datetimeOfPickup.strftime('%d %b %Y %I:%M %p ')} {way}"\
        f"\n\t   {self._hotel} Number of Pax: {self._numofPassangers}"\
        f"\n\t   Assigned: {driver}"

class TransportServices:
    
    def __init__(self):
        
        self._driverList  = []
        self._hotelsList  = []
        self._bookingList = []
    
    def searchHotel(self,hotelname):
        
        for h in self._hotelsList:
            if h.hotelName == hotelname:
                return h
        return None
    
    def searchDriver(self,driverName,vehicleNumber):
        for drivers in self._driverList:
            if drivers.driverName == driverName and drivers.vehicleNum == vehicleNumber:
                return drivers
        return None
    
    def searchBooking(self,pickupDateTime,fromAirport,hotelName):
        for bookings in self._bookingList:
            if bookings.datetimeOfPickup == pickupDateTime and bookings.fromAirport == fromAirport and bookings.hotel.hotelName == hotelName:
                return bookings
        return None
    
    def addHotel(self,hotel):
        if not self.searchHotel(hotel.hotelName):
            self._hotelsList.append(hotel)
            return True
        return False
    
    def addDrivers(self,driver):
        if not self.searchDriver(driver.driverName,driver.vehicleNum):
            self._driverList.append(driver)
            return True
        return False
    
    def addBookings(self,booking):
        if not self.searchBooking(booking.datetimeOfPickup,booking.fromAirport,booking.hotel.hotelName):
            self._bookingList.append(booking)
            return True
        return False
    
    def removeDriver(self,name,vehicleNumber):
        if not self.searchDriver(name,vehicleNumber):
            return False
        else:
            self._driverList.remove(self.searchDriver(name,vehicleNumber))
            return True
    
    def removeBooking(self,pickupDateTime,fromAirport,hotelName):
        if not self.searchBooking(pickupDateTime,fromAirport,hotelName):
            return False
        else:
            self._bookingList.remove(self.searchBooking(pickupDateTime,fromAirport,hotelName))
            return True
    
    def assignDriver(self,pickupDateTime,fromAirport,hotelName,driverName,vehicleNumber):
        if not self.searchDriver(driverName,vehicleNumber) or not self.searchBooking(pickupDateTime,fromAirport,hotelName):
            return False
        elif self.searchBooking(pickupDateTime,fromAirport,hotelName).driver:
            return False
        else:
            self.searchBooking(pickupDateTime,fromAirport,hotelName).driver = self.searchDriver(driverName,vehicleNumber)
            return True
    
    def driverStr(self):
        s=''
        for drivers in self._driverList:
            s+=f'{drivers}\n'
        return s
            
    def bookingStr(self):
        s=''
        for booking in self._bookingList:
            s+=f"{booking}\n"
        return s

    def hotelStr(self):
        s = ''
        for hotel in self._hotelsList:
            s+=f"{hotel}\n"
        return s
    
def main():
    
    h1=Hotel('Zenite','1, One Road P111')
    h2=Hotel('Young Lodge','2, Two Road P222')
    
    d1=Driver('Alan','A1110')
    d2=Driver('Betty','B2222')
    d3=Driver('Charlie','C3333')
    
    b1=Booking(datetime(2019,12,1,11,45),True,4,h1)
    b2=Booking(datetime(2019,12,2,9,45),False,1,h1)
    
    t=TransportServices()
    t.addDrivers(d1)
    t.addDrivers(d2)
    t.addDrivers(d3)
    t.addHotel(h1)
    t.addHotel(h2)
    t.addBookings(b1)
    t.addBookings(b2)
    
    
    
    print(t.driverStr())
    print(t.hotelStr())
    print(t.bookingStr())
    t.searchDriver('Alan','A1110').vehicleNum = 'A1111'
    
    print(t.searchBooking(datetime(2019,12,1,11,45),True,'Zenite'))
    
    print(t.assignDriver(datetime(2019,12,1,11,45),True,'Zenite','Alan','A1111'))
    print(t.bookingStr())
    print(t.driverStr())

main()
```

```python
from abc import abstractmethod , ABC
from datetime import datetime , timedelta

class CoOperativeException(Exception):
    """
    Exception for Coperation
    """

class Member:
    
    _nextID = 1
    
    def __init__(self,name,contact,dob,monthlyIncome=None):
        
        self._name = name
        self._contact = contact
        self._dob = dob
        self._monthlyIncome = monthlyIncome
        self._memberID = Member._nextID
        Member._nextID+=1
    
    @property
    def memberID(self):
        return self._memberID
    @property
    def monthlyIncome(self):
        return self._monthlyIncome
    
    def age(self,date):
        return (date-self._dob).days // 365
    
    def __str__(self):
        if not self._monthlyIncome:
            MI = 'Undisclosed'
        else:
            MI = f"${self._monthlyIncome}"
        return f"Id: {self._memberID:<3}Name: {self._name}\tContact: {self._contact:<10} MonthlyIncome: {MI}  Date Of Birth:"\
    f" {self._dob.strftime('%d %b %Y')}"
    

class Course(ABC):
    
    _baseCancellationPenaltyRate = 0.30
    
    def __init__(self,courseCode,title,description,fees):
        
        self._courseCode = courseCode
        self._title = title
        self._description = description
        self._fees = fees
    
    @classmethod
    def getbaseCancellationPenaltyRate(cls):
        return cls._baseCancellationPenaltyRate
    @classmethod
    def setbaseCancellationPenaltyRate(cls,newValue):
        cls._baseCancellationPenaltyRate = newValue
    
    @property
    def courseCode(self):
        return self._courseCode
    @property
    def fees(self):
        return self._fees
    
    @abstractmethod
    def getSubsidy(self,income,age):
        pass
    
    def cancellationPenaltyRate(self,days):        
        if days>7:
            return 0
        elif 4<= days <=7:
            return type(self)._baseCancellationPenaltyRate * 0.5
        else:
            return type(self)._baseCancellationPenaltyRate
    
    def cancellationPenalty(self,monthlyIncome,age,days):
        return (self._fees - self.getSubsidy(monthlyIncome,age)) * self.cancellationPenaltyRate(days)
    
    def __str__(self):
        return f"Course Code: {self._courseCode} Title: {self._title} Fees: ${self._fees}"\
    f"\n\tDescription: {self._description}"
    
class VocationalCourse(Course):
    
    _cancellationPenaltyAddRate = 0.1
    _specialisedIndustries = ['Engineering','Electronics','Computing']
    
    def __init__(self,courseCode,title,description,fees,industry):
        self._industry = industry
        super().__init__(courseCode,title,description,fees)
    
    def getSubsidy(self,income,age):
        
        if not income:
            rate = 0 
        else:
            if age >=60:
                rate = 0.95
            elif 55<= age < 60 :
                rate =  0.90
            else:
                rate = 0.80
        if rate == 0:
            return rate
        else:
            if self._industry in type(self)._specialisedIndustries:
                return self.fees * (rate - 0.1)
            else:
                return self.fees * rate
    
    def cancellationPenaltyRate(self,days):
        return super().cancellationPenaltyRate(days) + type(self)._cancellationPenaltyAddRate
    
    def __str__(self):
        return f"Vocational {super().__str__()} Industry: {self._industry}"
            
class NonVocationalCourse(Course):
    
    def getSubsidy(self,income,age):
        if not income:
            return 0
        else:
            if income<=1200 and age>=55:
                return self.fees * 0.50
            return 0
    def __str__(self):
        return f"Non Vocational {super().__str__()}"
            
class ScheduledCourse:
    
    def __init__(self,course,scheduledDate):
        self._course = course
        self._scheduledDate = scheduledDate
        self._participatingList = []
    
    @property
    def course(self):
        return self._course
    @property
    def scheduledDate(self):
        return self._scheduledDate
    @property
    def participatingList(self):
        return self._participatingList
    
    def searchParticipant(self,memeberID):
        for memeber in self._participatingList:
            if memeber.memberID == memeberID:
                return memeber
        return None
    
    def addParticipant(self,participant):
        if not self.searchParticipant(participant.memberID):
            self._participatingList.append(participant)
            return True
        return False
    
    def removeParticipant(self,participant):
        if self.searchParticipant(participant.memberID):
            self._participatingList.remove(participant)
            return True
        return False
    
    def __str__(self):
        s=''
        count = 0
        for participants in self._participatingList:
            s+=f"{participants}  Age: {participants.age(self._scheduledDate)}\n"
            count+=1
        return f"Scheduled Date: {self._scheduledDate.strftime('%d %b %Y')}\n"\
        f"{self._course}\n"\
        f"Participant List:\n{s}"\
        f"Number Of Participants: {count}"

class Cooperative:
    
    def __init__(self):
        
        self._memberDic = {}
        self._courseDic = {}
        self._scheduledcourseDateDic ={}
    
    def getMember(self,memberID):
        try:
            return self._memberDic[memberID]
        except KeyError:
            return None
        
    def addMember(self,member):
        if not self.getMember(member.memberID):
            self._memberDic[member.memberID]=member
            return True
        else:
            raise CoOperativeException(f"Duplicate member id {member.memberID}")
    
    def getCourse(self,courseCode):
        try:
            return self._courseDic[courseCode]
        except KeyError:
            return None
    
    def addCourse(self,course):
        if not self.getCourse(course.courseCode):
            self._courseDic[course.courseCode] = course
            return True
        else:
            raise CoOperativeException(f'Duplicate course code {course.courseCode}')
        
    
    def getScheduledCourse(self,courseCode,scheduleDate):
        
        try:
            return self._scheduledcourseDateDic[(courseCode,scheduleDate)]
        except KeyError:
            return None
    
    def addScheduledCourse(self,scheduledCourse):
        
        if not self.getCourse(scheduledCourse.course.courseCode):
            raise CoOperativeException(f'No course with code {scheduledCourse.course.courseCode}')
        elif self.getScheduledCourse(scheduledCourse.course.courseCode,scheduledCourse.scheduledDate):
            date=scheduledCourse.scheduledDate.strftime('%d %b %Y')
            raise CoOperativeException(f"Duplicate schedule. Course with code {scheduledCourse.course.courseCode} has already been scheduled on {date}")           
        else:
            self._scheduledcourseDateDic[(scheduledCourse.course.courseCode,scheduledCourse.scheduledDate)] = scheduledCourse
            return True
    
    def getMemberCourseAndScheduledCourse(self,memberID,courseCode,scheduleDate):
        if not self.getMember(memberID):
            raise CoOperativeException(f'No member with member id {memberID}')
        elif not self.getCourse(courseCode):
            raise CoOperativeException(f'No course with code {courseCode}')
        elif not self.getScheduledCourse(courseCode,scheduleDate):
            date=scheduleDate.strftime('%d %b %Y')
            raise CoOperativeException(f"No schedule for course with code {courseCode} on {date}")
        else:
            return self.getMember(memberID),self.getCourse(courseCode),self.getScheduledCourse(courseCode,scheduleDate)
    
    def enroll(self,memberID,courseCode,scheduledDate):
        member,course,scheduledcourse = self.getMemberCourseAndScheduledCourse(memberID,courseCode,scheduledDate)
        scheduledcourse.participatingList.append(member)
        return True
    
    def cancelEnrollment(self,memberID,courseCode,scheduleDate,cancelDate):
        member,course,scheduledcourse = self.getMemberCourseAndScheduledCourse(memberID,courseCode,scheduleDate)
        if not scheduledcourse.removeParticipant(member):
            return -1
        return course.cancellationPenalty(member.monthlyIncome,member.age(scheduleDate),(scheduleDate-cancelDate).days) 
    
    def membersStr(self):
        s=""
        for members in self._memberDic:
            s+=f"{self._memberDic[members]}\n\n"
        s=s.strip()
        return s
    def coursesStr(self):
        s=""
        for course in self._courseDic:
            s+=f"{self._courseDic[course]}\n\n"
        s=s.strip()
        return s
    def scheduledCoursesStr(self):
        s=""
        for course in self._scheduledcourseDateDic:
            s+=f"{self._scheduledcourseDateDic[course]}\n\n"
        s=s.strip()
        return s        
    
def menu():
    print("""Menu
1. List Members
2. List Courses
3. List Schedules
4. Add Member
5. Add Course
6. Add Schedule
7. Enroll Member
8. Cancel Enrollment
0. Exit""")
    
    try:
        option = int(input('Enter choice: '))
        if option<0 or option>8:
            raise CoOperativeException('Please Enter Valid Input 0-8!')
        return option
    except ValueError:
        raise CoOperativeException('Invalid Input !')
        
def addMember(co):
    
    name = input('Enter Name : ')
    while True:
        try:
            contact = int(input('Enter Contact : '))
            break
        except ValueError:
            print('Invalid Input ! Please Enter Valid Integer Value')
    while True:
        try:
            dob = input('Enter Date Of Birth in dd/mm/yy  format : ').strip()
            dob = datetime.strptime(dob,'%d/%m/%Y')
            break
        except ValueError:
            print('Invalid Input! Please enter date in d/m/yyyy')
    while True:
        try:
            mincome = int(input('Enter Monthly Income to nearest whole number or <Enter> for not disclosing : '))
            if mincome == '':
                MI = None
                
            else:
                MI = mincome
            break
        except ValueError:
            print('Invalid! Please Enter <Enter> or Integer Value ')
    
    try:
        co.addMember(Member(name,contact,dob,MI))
        print('Add Member Operation Successfull')
    except CoOperativeException as e:
        print(e)
        print('Add Member Operation Unsuccessfull ! ')
        
def addCourse(co):
    
    course = input('Enter V for vocational course & N for non-vocational courses : ').upper()
    while course!='V' and course!='N':
        print('Type of Course should be N or V.')
        course = input('Enter V for vocational course & N for non-vocational courses : ').upper()
    code = input('Enter Course Code : ').strip()
    title = input('Enter Title ').title()
    descrp = input('Enter Description : ')
    while True:
        try:
            fees = int(input('Enter Fees : '))
            break
        except ValueError:
            print('Please enter whole number')
    if course=='V':
        industry = input('Enter Industry : ')
        try:
            co.addCourse(VocationalCourse(code,title,descrp,fees,industry))
            print('Add course is Successful')
        except CoOperativeException as e :
            print(e)
    else:
        try:
            co.addCourse(NonVocationalCourse(code,title,descrp,fees))
            print('Add course is Successful')
        except CoOperativeException as e :
            print(e)
            
def addScheduleCOurse(co):
    
    code = input('Enter Course Code : ').strip()
    while True:
        try:
            sch_date = input('Enter Date Of Birth in dd/mm/yy  format : ').strip()
            sch_date = datetime.strptime(sch_date,'%d/%m/%Y')
            break
        except ValueError:
            print('Invalid Input! Please enter date in d/m/yyyy')
    try:
        if not co.getCourse(code):
            print(f"Invalid Course Code. No course with code {code}")
        else:
            co.addScheduledCourse(ScheduledCourse(co.getCourse(code),sch_date))
    except CoOperativeException as e :
        print(e)
    
def enroll(co):
    
    iD = input("Enter members ID: ")
    code = input("Enter Course Code: ")
    while True:
        try:
            sch_date = input('Enter Date Of Birth in dd/mm/yy  format : ').strip()
            sch_date = datetime.strptime(sch_date,'%d/%m/%Y')
            break
        except ValueError:
            print('Invalid Input! Please enter date in d/m/yyyy')
    try:
        co.enroll(iD,code,sch_date)
    except CoOperativeException as e:
        print(e)
    else:
        course = co.getCourse(code)
        member = co.getMember(iD)
        subs = co.getSubsidy(member.monthlyIncome,member.age(sch_date))
        print(member)
        print(f"Course Fee: ${course.fees} Subsidy: ${subs} Payment: ${course.fees-subs}")
        print("Enroll Operation is Successfull")
        

def cancelEnrollment(co):
    
    iD = input("Enter members ID: ")
    code = input("Enter Course Code: ")
    while True:
        try:
            sch_date = input('Enter Date Of Birth in dd/mm/yy  format : ').strip()
            canceldate = input('Enter Cancel Date in dd/mm/yy  format : ').strip()
            sch_date = datetime.strptime(sch_date,'%d/%m/%Y')
            canceldate = datetime.strptime(canceldate,'%d/%m/%Y')
            break
        except ValueError:
            print('Invalid Input! Please enter date in d/m/yyyy')

    try:
        amount = co.cancelEnrollment(iD,code,sch_date,canceldate)
    except CoOperativeException as e:
        print(e)
    else:
        if amount == -1:
            print(f'Member of {iD} is not a participant of course with code{code}')
        else:
            course = co.getCourse(code)
            member = co.getMember(iD)
            subs = co.getSubsidy(member.monthlyIncome,member.age(sch_date))
            days = (sch_date-canceldate).days
            print(f"Cancellation rate for days {days} {course.cancellationPenaltyRate(days)}")
            print(member)
            print(f"Course Fee: ${course.fees} Subsidy: ${subs} Payment: ${course.fees-subs}")
            print(f"Penalty Rate: {course.cancellationPenaltyRate(days)}% Penalty Amount: ${amount}")
            print("Enroll Operation is Successfull")
    
def main():
    
    m1 = Member('Helen',111111,datetime(1964,1,1),10000)
    m2 = Member('Irene',222222,datetime(1949,10,1))
    vc=VocationalCourse('E012','Engeninering In Business','Developpment& Implementation of business Solution',1000,'Engineering')
    nvc= NonVocationalCourse('C001','Gluten-Free Diet',"How to Start a GLuten-Free Diet",100)
    s1=ScheduledCourse(nvc,datetime(2019,10,12))
    s2=ScheduledCourse(vc,datetime(2019,10,12))
    
    co=Cooperative()
    co.addMember(m1)
    co.addMember(m2)
    co.addCourse(vc)
    co.addCourse(nvc)
    co.addScheduledCourse(s1)
    co.addScheduledCourse(s2)
    
    while True:
        
        try:
            option = menu()
        except CoOperativeException as e:
            print(e)
        else:
            if option == 0:
                print("Application Ended!")
                break
            elif option == 1:
                print(co.membersStr())
            elif option == 2:
                print(co.coursesStr())
            elif option == 3:
                print(co.scheduledCoursesStr())
            elif option == 4:
                addMember(co)
            elif option == 5 :
                addCourse(co)
            elif option == 6:
                addScheduleCOurse(co)
            elif option==7:
                enroll(co)
            elif option==8:
                cancelEnrollment(co)                
                
main()
```
