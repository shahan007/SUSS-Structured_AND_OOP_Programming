```python
from abc import ABC , abstractmethod

class Item(ABC):
    
    _administrativeCharge = 30
    
    def __init__(self,isbn,title,price,yearPublication,authors=[]):
        
        self._isbn = isbn
        self._title = title.title()
        self._price = price
        self._yearPublication = yearPublication
        self._authors = authors
    
    @classmethod
    def administrativeCharge(cls):
        return cls._administrativeCharge
    @classmethod
    def setadministrativeCharge(cls,newValue):
        cls._administrativeCharge = newValue
    
    @property
    def isbn(self):
        return self._isbn
    @property
    def title(self):
        return self._title
    @property
    def price(self):
        return self._price
    @price.setter
    def price(self,newValue):
        self._price = newValue
    @property
    def authors(self):
        return ', '.join(self._authors)
        
    @property
    def yearPublication(self):
        return self._yearPublication
    
    @abstractmethod
    def getLateCharges(self,noOfDays):
        pass
    @abstractmethod
    def lossSurChargeRate(self):
        pass
    
    def lossPenalty(self):
        return self._price + (self.lossSurChargeRate()*self._price) + type(self).administrativeCharge()
    
    def matchKeyWords(self,listOfKeyWords):
        kw=[words.lower() for words in listOfKeyWords]
        strl=self._title.strip().lower().split()
#         kw.sort()
#         strl.sort()
        return all(items in strl for items in kw)

    
    def isWrittenBy(self,author):
        return author.lower().strip() in self.authors.lower()
    
    def __str__(self):
        
        athr=None
        if len(self._authors)>1:
            athr = 'Authors'
        else:
            athr="Author"
        
        return f"ISBN: {self._isbn} Title: {self._title} Price: ${self._price:0.2f} "\
    f"\n{athr}: {self.authors} Year Published: {self._yearPublication}"
        
        
class Book(Item):
    
    _lossSurChargeRate = 0.2
    
    def __init__(self,isbn,title,price,yearPublication,edition,authors=[]):
        self._edition = edition
        super().__init__(isbn,title,price,yearPublication,authors)
    
    @classmethod
    def getlossSurChargeRate(cls):
        return cls._lossSurChargeRate
    
    @classmethod
    def setlossSurChargeRate(cls,newValue):
        cls._lossSurChargeRate = newValue
        
    @property
    def edition(self):
        return self._edition
    def getLateCharges(self,noOfDays):
        
        charge=None
        
        if noOfDays<=7:
            return round((0.20 * noOfDays),2)
        elif  7< noOfDays < 15:
            return round(((0.20 * 7) + (noOfDays-7)*0.5),2)
        else:
            charge = (0.20 * 7) + (7*0.5) + (noOfDays-14)
            if charge > self.lossPenalty():
                return self.lossPenalty()
            return round(charge,2)
                
    
    def lossSurChargeRate(self):
        return type(self).getlossSurChargeRate()
    
    def __str__(self):
        return f"{super().__str__()} Edition: {self._edition}"
            
class Library:
    
    def __init__(self):
        self._items = {}
        
    def searchItem(self,isbn):
        try:
            return self._items[isbn]
        except KeyError:
            return None
    
    def addItem(self,item):
        
        if self.searchItem(item.isbn)!=None:
            return False

        else:
            self._items[item.isbn] = item
            return True
    
    def removeItem(self,isbn):
        try:
            self._items.pop(isbn)
            return True
        except KeyError:
            return False

    @classmethod
    def __showResult(cls,collectionItems):
        no=1
        s=""
        for items in collectionItems:
            s+=f'Item {no}:\n{items}\n\n'
            no+=1
        return s
    
    def itemsByAuthor(self,author):
        
        collection=[]
        for items in self._items.values():
            if items.isWrittenBy(author):
                collection.append(items)
            else:
                continue
        return type(self).__showResult(collection)
    
    def itemsWithTitle(self,title):
        collection=[]
        for items in self._items.values():
            if items.title==title.title():
                collection.append(items)
            else:
                continue
        return type(self).__showResult(collection)
    
    def itemsWithKeywords(self,listOfKeyWords):
        collection=[]
        for items in self._items.values():
            if items.matchKeyWords(listOfKeyWords):
                collection.append(items)
            else:
                continue
        return type(self).__showResult(collection)
    
    def listItems(self):
        return type(self).__showResult(list(self._items.values()))         

def main():
    
    l=Library()
    b1=Book(9781234567890,'Python For Beginners',29.99,'2019',2,['Tom Smart','Alice','Alanis'])
    b2=Book(9781234567891,'Fun With Python',10.00,'2018',1,['Tom Smart'])
    l.addItem(b1)
    l.addItem(b2)
    print(l.listItems())
    print(l.itemsByAuthor('Tom Smart'))
    print(l.itemsWithKeywords(['fun','python']))
    print(l.itemsWithKeywords(['python']))        
    l.removeItem(9781234567890)    
    print(l.listItems())
    print("\nCharges:")
    print(l.searchItem(9781234567891).getLateCharges(1))
    print(l.searchItem(9781234567891).getLateCharges(8))
    print(l.searchItem(9781234567891).getLateCharges(19))
    print(l.searchItem(9781234567891).getLateCharges(100))

main()      
```

```python
class TravelException(Exception):
    """
    Travel Exception
    """

class InsufficientAmountException(TravelException):
    """
    InsufficientAmountException
    """
    
class InputError(TravelException):
    """
    Input Exception
    """

class TravelPass:
    
    _costOfATrip = 1.50
    _numOfTrips = 10
    _refundPerTrip = 1.20
    _adminCharge = 5
    
    _id = 1
    
    
    def __init__(self,amount):
        
        if amount<TravelPass.costOfTravelPass():
            raise InsufficientAmountException(f'Minimum amount to buy a pass is ${TravelPass.costOfTravelPass()}')
        else:
            self._amount = amount
        
        self._travelID = TravelPass._id
        TravelPass._id +=1
        self._unconsumedTrips = type(self)._numOfTrips
        
    @property
    def unconsumedTrips(self):
        return self._unconsumedTrips
    
    def makeTrip(self):
        if self._unconsumedTrips == 0:
            raise TravelException("Travel pass has 0 trips left")
        else:
            self._unconsumedTrips-=1
    def getRefund(self):
        if self._unconsumedTrips==0:
            raise TravelException('There is no trip refund')
        else:
            refund_amount = type(self).getRefundPerTrip() * self._unconsumedTrips
            self._unconsumedTrips = 0
            return refund_amount
    
    
    def topUp(self,amount):
        if amount<TravelPass.getCostPerUnit():
            raise InsufficientAmountException(f'Minimum amount to buy a pass is ${TravelPass.getCostPerUnit()}')
        else:
            old = self._unconsumedTrips
            change = amount -  TravelPass.getCostPerUnit()
            self._unconsumedTrips+=TravelPass.getNumberOfTripsPerUnit()
            return f"Before top up: Travel Pass: {self._travelID:07d} Number of trips left: {old}"\
        f"\nAfter 1 unit top up: Travel Pass: {self._travelID:07d} Number of trips left:{self._unconsumedTrips} "\
        f'\nChange=${change:0.2f}'    
    
    @classmethod
    def getNumberOfTripsPerUnit(cls):
        return cls._numOfTrips
    @classmethod
    def getCostPerUnit(cls):
        return cls._costOfATrip * cls._numOfTrips
    @classmethod
    def costOfTravelPass(cls):
        return cls._adminCharge + (cls._numOfTrips * cls._costOfATrip)
    
    @classmethod
    def getRefundPerTrip(cls):
        return cls._refundPerTrip
    
    def __str__(self):
        return f"Travel Pass: {self._travelID:07d} Number of trips left: {self._unconsumedTrips}"
    
class Passanger:
    
    def __init__(self,passangerID,travelPass=None):
        
        self._passangerID = passangerID
        self._travelPass = travelPass
        
    @property
    def travelPass(self):
        return self._travelPass
    
    def hasTravelPass(self):
        if not self._travelPass:
            return False
        else:
            return True
    
    def getTravelPass(self,amount):
        if self.hasTravelPass():
            if self._travelPass.unconsumedTrips==0:
                self._travelPass = TravelPass(amount)
            else:
                raise TravelException(f'Please use existing {self._travelPass.unconsumedTrips} trips in current pass or get a refund first')
        else:
            self._travelPass = TravelPass(amount)
    def topUp(self,amount):
        if not self._travelPass:
            raise TravelException('No travel pass! Please get a travel pass')
        else:
            return self._travelPass.topUp(amount)
    
    def refund(self):
        if not self._travelPass:
            raise TravelException('No travel pass! Please get a travel pass')
        else:
            refund=self._travelPass.getRefund()
            self._travelPass=None
            return refund
    
    def makeTrip(self):
        if not self._travelPass:
            raise TravelException('No travel pass! Please get a travel pass')
        else:
            self._travelPass.makeTrip()
    
    def __str__(self):
        return f"Passenger Id: {self._passangerID} {self._travelPass}"
        
def getMenuOption():
    print("""Menu
1. Get Travel Card
2. Get refund
3. Make a trip
4. Top up Travel Pass
0. Exit
    """)
    try:
        option=int(input('Enter choice:'))
        if option<0 or option>4:
            raise InputError(f'Invalid Choice: {option}')
        return option
    except ValueError:
        raise InputError('Error in choice. Please enter a whole number 0 to 4.')

def getAmount():
    
    try:
        amount=float(input("Enter Amount: "))
        if amount<0:
            raise InputError('Error in amount. Please enter a positive decimal number')
        return amount
    except ValueError:
        raise InputError('Error in amount. Please enter a positive decimal number')
        
def getTravelPass(p):
    
    if not p.hasTravelPass():
        amount=getAmount()
        p.getTravelPass(amount)
        return f"Charge=${amount-p.travelPass.costOfTravelPass():0.2f}"
    else:
        if p.travelPass.unconsumedTrips != 0:
            raise TravelException(f'Please use existing {p.travelPass.unconsumedTrips} trips in current pass or get a refund first')
        else:
            amount=getAmount()
            top_UP= input('To top up instead, enter T. Enter any key to purchase new pass:').lower()
            if top_UP[0]=='t':
                return p.topUp(amount)
            else:
                p.getTravelPass(amount)
                return f"Charge=${amount-p.travelPass.costOfTravelPass():0.2f}"


def topUp(p):
        amount=getAmount()
        return p.topUp(amount)

def main():
    
    passanger1=Passanger('T1234X')
    
    while True:
        print(passanger1)
        try:
            option=getMenuOption()
        except InputError as e :
            print(e)
        else:
            if option==0:
                print('Application Ended')
                break
            elif option==1:
                try:
                    print(getTravelPass(passanger1))
                    print(passanger1)
                    print('Get Travel Card operation is successful.')
                except InputError as e:
                    print(e)
                except InsufficientAmountException as e:
                    print('Insufficient amount to Get Travel Card operation.')
                    print(f"Reason : {e}")
                except TravelException as e :
                    print('Get Travel Card operation is unsuccessful.')
                    print(f'Reason: {e}')
            elif option==2:
                try:
                    print(f'Refund amount at ${TravelPass.getRefundPerTrip()} per trip: ${passanger1.refund()}')
                    print(passanger1)
                    print('Get refund operation is successful.')
                except TravelException as e :
                    print('Get refund operation is unsuccessful.')
                    print(f'Reason: {e}')
                
            elif option==3:
                try:
                    passanger1.makeTrip()
                    print(passanger1)
                    print('Make a trip operation is successful.')
                except TravelException as e:
                    print(f'Make a trip operation is unsuccessful.')
                    print(f'Reason: {e}')                    
            elif option==4:
                try:
                    print(topUp(passanger1))
                    print(passanger1)
                    print('Top up Travel Pass operation is successful.')
                except InputError as e:
                    print(e)
                except InsufficientAmountException as e:
                    print('Insufficient amount to Top up Travel Pass operation')
                    print(f"Reason : {e}")
                except TravelException as e:
                    print('Top up Travel Pass operation is unsuccessful.')
                    print(f'Reason: {e}')
       
        print('\n')
        
main()
```

```python
from tkinter import *
from tkinter.scrolledtext import ScrolledText
from datetime import datetime , timedelta

class CarRental(Frame):
    
    def __init__(self):
        #----------Creating Window-------------------
        Frame.__init__(self)
        self.master.title('Car Rental Enquiry')
        self.master.geometry('581x325')
        #------------Creating Components-------------
        self._classLabel = Label(self,text='Class of Car (A,B or C):                  ')
        self._sdLabel = Label(self,text='Start Date of Rental (dd/mm/yy):')
        self._edLabel = Label(self,text='End Date of Rental (dd/mm/yy):  ')
        self._classEntry = Entry(self)
        self._sdEntry = Entry(self)
        self._edEntry = Entry(self)
        self._enquireButton = Button(self,text='Enquire',width = 10,command=self.enquire)
        self._clearButton = Button(self,text='Clear Data',width =10,command=self.clear)
        self._txtBOX = Text(self,width=72,height=5)
        self._txtBOX.insert(END,"\t\tClass:          {0:<10}{1:<10}{2}\n\n".format('A','B',"C"))
        self._txtBOX.insert(END,"\t\tWeekDay Rate: ${0:<10}${1:<10}${2}\n\n".format(100,150,250))
        self._txtBOX.insert(END,"\t\tWeekend Rate: ${0:<10}${1:<10}${2}\n\n".format(120,180,320))
        self._scrxltxt = ScrolledText(self,width=70,height=8)
        self._scrxltxt.configure(state=DISABLED)
        #----------------Placing Componenets--------------
        self._txtBOX.grid(row=0,column=0,columnspan=1,sticky=W)
        self._classLabel.grid(row=1,column=0,pady=3,sticky=E+W)
        self._classEntry.grid(row=1,column=0,sticky=E,padx=(0,70))
        self._sdLabel.grid(row=2,column=0)
        self._sdEntry.grid(row=2,column=0,sticky=E,padx=(0,70))
        self._edLabel.grid(row=3,column=0)
        self._edEntry.grid(row=3,column=0,sticky=E,padx=(0,70))
        self._enquireButton.grid(row=4,column=0,pady=7,sticky=W,padx=240)
        self._clearButton.grid(row=4,column=0,sticky=E,padx=(0,175))
        self._scrxltxt.grid(row=6,column=0,columnspan=15,sticky=W)        
        self.grid()
        
    def enquire(self):
        self._scrxltxt.configure(state=NORMAL)

        clas= self._classEntry.get()
        ed = datetime.strptime(self._edEntry.get(),'%d/%m/%Y')
        sd = datetime.strptime(self._sdEntry.get(),'%d/%m/%Y')

        if clas!='A' and clas!='B' and clas!='C':
            self._scrxltxt.insert(END,'Invalid Car Class\n')
        else:
            if ed<sd:
                self._scrxltxt.insert(END,'End Date cannot be earlier\n')
            else:
                if clas=='A':
                    weekday_rate = 100
                    weekend_rate = 120
                elif clas=='B':
                    weekday_rate = 150
                    weekend_rate = 180
                else:
                    weekday_rate = 250
                    weekend_rate = 320
                    
                    
                difference = (ed - sd).days
                no_of_weekdays = 0
                no_of_weekdends = 0
                weekday_money = 0
                weekdend_money = 0
                for day in range(difference+1):
                    sd+=timedelta(days=day)
                    if 0<sd.isoweekday()<6:
                        no_of_weekdays+=1
                        weekday_money+=weekday_rate
                    else:
                        no_of_weekdends+=1
                        weekdend_money+=weekend_rate
                    sd-=timedelta(days=day)
                TotalCost = weekday_money + weekdend_money
                sd = sd.strftime('%a,%d %b %Y')
                ed = ed.strftime('%a,%d %b %Y')
                
                self._scrxltxt.insert(END,f'Car Class {clas}\n')
                self._scrxltxt.insert(END,f'Start Date: {sd}\tEnd Date:{ed}\n')
                self._scrxltxt.insert(END,f'Number of Weekdays: {no_of_weekdays}\tWeekday Rate: ${weekday_rate}\tCost: ${weekday_money}\n')
                self._scrxltxt.insert(END,f'Number of Weekdends: {no_of_weekdends}\tWeekend Rate: ${weekend_rate}\tCost: ${weekdend_money}\n')
                self._scrxltxt.insert(END,f'Total Cost: ${TotalCost}\n\n')
        self._scrxltxt.configure(state=DISABLED)
            
    
    def clear(self):
        
        self._scrxltxt.configure(state=NORMAL)
        lis = [self._classEntry,self._sdEntry,self._edEntry]
        
        for x in lis:
            x.delete(0,END)
        self._scrxltxt.delete(1.0,END)
        self._scrxltxt.configure(state=DISABLED)

def main():
    car = CarRental()
    car.mainloop()

main()
```
