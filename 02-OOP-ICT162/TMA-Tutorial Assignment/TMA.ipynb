<b>Requirements:</b> <br><br>
write an application for a company that uses a subscription-based business model to loan out clothes such as office and party wear to female clients.<br>
A subscription costs $69 per 4 weeks.<br>
A client can make unlimited loan requests during her subscription, as long as she has no previous loan that has not been returned.<br>
There can be only one piece of clothes in each loan request.<br>
The chosen piece of clothes takes a day to be delivered.<br>
For example, if a piece of clothes is selected on Wednesday (say, 4th September 2019), then it will arrive at the
address of the client the following day, that is, Thursday, 5th 2019
The subscription is automatically renewed. To cancel subscription, a client must
o not have any loan that is unreturned,
o pay up any outstanding balance, and
o call to cancel before week 4 of each 4-week cycle

```python
from datetime import timedelta,datetime

class Client:
    
    
    def __init__(self,clientid,name,address):
        self._clientid = clientid
        self._name = name
        self._address = address
        self._balance = 0.0
        
    @property  #Getter for Client's id        
    def clientid(self):
        return self._clientid
    @property  #Getter for Client's name
    def name(self):
        return self._name
    @property  #Getter for CLient's address
    def address(self):
        return self._address
    @address.setter #Setter for Client's address to modify data when needed
    def address(self,newAddress):
        self._address = newAddress
    @property  #Getter for Client's balance
    def balance(self):
        return self._balance
    
    def addPayment(self,amount):
        """
        Method adds amount to client's balance and return True if amount is positive
        else returns False
        
        """
        if amount>0:
            self._balance+=amount
            return True
        return False
    
    def makePayment(self,amount):
        """
        Method deducts amount from client's balance and return True if amount is positive
        else returns False
        
        """
        if amount>0:
            self._balance-=amount
            return True
        return False
    
    def __str__(self):
        return f"Id: {self._clientid} Name: {self._name} Address: {self._address} Payment due: ${self._balance}"

class Clothes:
    
    def __init__(self,clothesid,brand,size,price):
        self._clothesid = clothesid
        self._brand = brand
        self._size = size
        self._price = price
        self._onLoan = False
    
    @property   #Getter for Clothes id
    def clothesid(self):
        return self._clothesid
    @property   #Getter for Clothes brand
    def brand(self):
        return self._brand
    @property   #Getter for Clothes size
    def size(self):
        return self._size
    @property   #Getter for Clothes Price
    def price(self):
        return self._price
    @property   #Getter for Clothes Loan Status
    def onLoan(self):
        return self._onLoan
    @onLoan.setter  #Setter for Clothes Loan Status
    def onLoan(self,newValue):
        self._onLoan = newValue
    
    def __str__(self):
        
        status = None
        if self._onLoan:
            status='yes'
        else:
            status="no"
        return f"Id: {self._clothesid:<6} Brand: {self._brand:<10} Size: {self._size} Price: ${self._price} Out: {status}"


class Subscription():
    
    _nextId = 1
    _subscriptionRate = 69
    
    def __init__(self,startDate,client):
        self._subscriptionId = type(self)._nextId
        type(self)._nextId+=1    
        self._startDate = startDate
        self._client = client
        self._client.addPayment(type(self)._subscriptionRate)
        self._clothes = None
        self._deliveredDate = None
        self._returnedDate = None
    
    @property  #Getter for Subscriber's id
    def subscriptionId(self):
        return self._subscriptionId
    @property  #Getter for Subscriber
    def client(self):
        return self._client
    @property  #Getter for Subscriber's Loaned Clothes
    def clothes(self):
        return self._clothes
    @clothes.setter  #Setter for Subscriber's Loaned Clothes
    def clothes(self,newValue):
        self._clothes = newValue  
    @property   #Getter for Subscriber's Start Date of Subscription
    def startDate(self):
        return self._startDate
    @property   #Getter for Delivery Date of Clothes to Subscriber
    def deliveredDate(self):
        return self._deliveredDate        
    @deliveredDate.setter  #Setter for Delivery Date of Clothes to Subscriber
    def deliveredDate(self,newValue):
        self._deliveredDate = newValue
    @property  #Getter for Returned Date of Clothes from Subscriber
    def returnedDate(self):  
        return self._returnedDate                
    @returnedDate.setter  #Setter for Returned Date of Clothes from Subscriber
    def returnedDate(self,newValue):
        self._returnedDate = newValue
    @classmethod
    def getSubscriptionRate(cls):
        return cls._subscriptionRate
    
    def returnClothes(self,returnedDate):
        """
        Method returns boolean upon successful return of loaned clothe
        
        """
        
        if self._clothes == None :
            return False
        elif self._deliveredDate!=None and returnedDate<self._deliveredDate:
            return False
        else:
            self._clothes.onLoan = False
            self._clothes = None
            self._returnedDate = returnedDate
            return True
    def requestClothes(self,requestDate,clothes):
        """
        Method returns boolean upon successful request for loaning clothe
        
        """
        if self._clothes!=None :
            return False
        elif clothes.onLoan==True:
            return False
        elif self._returnedDate!=None and requestDate < self._startDate:
            return False
        elif self._returnedDate!=None and requestDate<self._returnedDate:
            return False
        else:
            self._deliveredDate = requestDate + timedelta(days=1)
            clothes.onLoan=True
            self._clothes = clothes
            return True
        
    def __str__(self):
        
        #Subscriber who have never loaned out clothes
        if self._clothes==None and self._returnedDate==None:
            s=datetime.strftime(self._startDate,"%a, %d %b %Y")
            return f"Subscription Id: {self._subscriptionId}  Start Date: {s}\n{self._client.__str__()}\nNo Outstanding Clothes."
        #Subscriber who has already returned their loaned clothes
        elif self._returnedDate!=None and self._clothes==None:
            s=datetime.strftime(self._startDate,"%a, %d %b %Y")
            r=datetime.strftime(self._returnedDate,"%a, %d %b %Y")
            return f"Subscription Id: {self._subscriptionId}  Start Date: {s}\n{self._client.__str__()}\nNo Outstanding Clothes.  Returned:{r}"
        #Subscriber who has loaned the clothe 
        elif self._clothes!=None:
            s=datetime.strftime(self._startDate,"%a, %d %b %Y")
            d=datetime.strftime(self._deliveredDate,"%a, %d %b %Y")
            return f"Subscription Id: {self._subscriptionId}  Start Date: {s}\n{self._client.__str__()}\nClothes loaned {self._clothes.__str__()} Delivery:{d}"


class Company():

    def __init__(self):
        self._clothes = []    #To store Clothes
        self._clients = {}    #To store CLients
        self._subscriptions = []  #To store Subscribers

    def searchClient(self,clientId):
        """
        Returns Client if present in Client's dictionary else returns None
        """
        if clientId in self._clients:
            return self._clients[clientId]
        return None

    def searchClothes(self,clothesid):
        """
        Returns Clothes if present in Clothes list else returns None
        """
        for clothes in self._clothes:
            if clothesid == clothes.clothesid:
                return clothes
        return None

    def searchSubscription(self,clientId):
        """
        Returns Subscriber if present in Subscriptions list else returns None
        """
        for subs in self._subscriptions:
            if clientId == subs.client.clientid:
                return subs
        return None

    def addClient(self,clientId,name,address):
        """
        Return True if adding of Client to Company is successful else returns False
        """
        if self.searchClient(clientId)==None:
            self._clients[clientId]=Client(clientId,name,address)
            return True
        else:
            return False

    def addClothes(self,clothe):
        """
        Return True if Clothe has been successfully added to the Company else returns False
        """
        if self.searchClothes(clothe.clothesid)==None:
            self._clothes.append(clothe)
            return True
        else:
            return False

    def addSubscription(self,startDate,clientId):
        """
        Return True if Client is successfully subscribed else returns False
        """
        if self.searchClient(clientId)!=None and self.searchSubscription(clientId)==None:
            self._subscriptions.append(Subscription(startDate,self.searchClient(clientId)))
            return True
        else:
            return False

    def cancelSubscription(self,clientId,cancelDate):
        """
        Returns True if Subscriber's subscription is successfully cancelled else returns False
        """
        #If client id is invalid return False
        if self.searchClient(clientId)==None:
            return False
        #f client is not subscribed return False
        elif self.searchSubscription(clientId)==None:
            return False
        elif self.searchSubscription(clientId).clothes!=None and self.searchSubscription(clientId).clothes.onLoan==True:
            return False
        #If client owes the company outstanding payment return False
        elif self.searchClient(clientId).balance>0:
            return False
        #If cancel date is before clients start date return False
        elif cancelDate<self.searchSubscription(clientId).startDate:
            return False
        #returns False if cancel date is not before week 4 of the 4 week subscription cycle
        elif ((cancelDate-self.searchSubscription(clientId).startDate).days)%28>21:
            return False
        else:
            for subs in self._subscriptions:
                if subs.client.clientid==clientId:
                    self._subscriptions.remove(subs)
                    return True

    def returnClothes(self,clientId,returnDate):
        """
        Returns True if Clothes have been successfully returned else returns False
        """
        if self.searchClient(clientId)==None or self.searchSubscription(clientId)==None:
            return False
        else:
            return self.searchSubscription(clientId).returnClothes(returnDate)

    def requestClothes(self,clientId,requestDate,clothes):
        """
        Returns True if Clothes have been successfully requested else returns False
        """
        if self.searchClient(clientId)==None:
            return False
        elif self.searchSubscription(clientId)==None:
            return False
        elif clothes.onLoan==True:
            return False
        else:
            return self.searchSubscription(clientId).requestClothes(requestDate,clothes)

    def clientStr(self):
        """
        Returns all the Clients in the Company
        """

        s=''
        for client in self._clients:
            s+=self._clients[client].__str__() + '\n'
        return s

    def clothesStr(self):
        """
        Return all the clothes in the Company
        """
        s=''
        for clothes in self._clothes:
            s+=clothes.__str__()+'\n'
        return s

    def availableClothesStr(self):
        """
        Return all the clothes that are available to be loaned out
        """

        s=""
        for clothes in self._clothes:
            if clothes.onLoan:
                continue
            else:
                s+=clothes.__str__()+'\n'
        return s

    def subscriptionStr(self):
        """
        Returns all the Subscribers registered with the Company
        """

        s=""
        for subs in self._subscriptions:
            s+=subs.__str__()+'\n'
        return s

    def __str__(self):
        return f"Clients:\n{self.clientStr()}\nClothes\n{self.clothesStr()}\nSubscription\n{self.subscriptionStr()}"



def test():
    
    print('-----------------------------------------')
    print('Running Test')
    print('-----------------------------------------\n\n')
          
    print('Creating Client')
    client1 = Client(123,"Janet Teo","21 Tampines West Road")
    print(client1)    
    print('')
    
    print('Creating Clothes')
    clothes1 = Clothes(104,'Juillet',6,214)
    clothes2 = Clothes(107,'Elliatt',7,237)
    print(clothes1)
    clothes2.onLoan = True
    print(clothes2)
    clothes2.onLoan = False
    print('')

    print('Creating Subscription')
    #Example 1: With no previous loan and no existing loan
    s=Subscription(datetime(2019,8,31),client1)
    print(s)
    print('')
    
    #Example 2: With a previous piece of clothes that is loaned, returned and no existing loan
    s=Subscription(datetime(2019,8,31),client1)
    s.requestClothes(datetime(2019,9,3),clothes2)
    s.returnClothes(datetime(2019,9,6))
    print(s)
    print('')
    
    #Example 3: With a loan of a piece of clothes
    s=Subscription(datetime(2019,8,31),client1)
    s.requestClothes(datetime(2019,9,7),clothes2)
    print(s)

    print('\n\n-----------------------------------------')
    print('Ending Test')
    print('-----------------------------------------\n\n\n')    

def main():

    c=Company()

    clothes1 = Clothes(34,'Charm',7,147)
    clothes2 = Clothes(91,'Elliatt',6,287)
    clothes3 = Clothes(104,'Juillet',6,214)
    clothes4 = Clothes(107,'Elliatt',7,237)
    clothes5 = Clothes(115,'Juillet',6,253)
    clothes6 = Clothes(123,'Suncoo',6,194)

    c.addClothes(clothes1)
    c.addClothes(clothes2)
    c.addClothes(clothes3)
    c.addClothes(clothes4)
    c.addClothes(clothes5)
    c.addClothes(clothes6)

    c.addClient(123,'Janet Teo',"21 Tampines West Road")
    c.addSubscription(datetime(2019,8,31),123)
    c.requestClothes(123,datetime(2019,9,7),clothes4)
    print(c)

test()
main()
```

<br><b>Requirements:</b> <br><br>
A delivery company provides both local and overseas deliveries which are either
normal delivery or express delivery.<br> Every sender must be registered.<br> To have an item
delivered, the sender calls up the company to request for the item to be picked up.<br> The
destinations of overseas deliveries are classified into 6 zones.

```python
from datetime import datetime , timedelta

class DeliveryException(Exception):
    """
    Raises Delivery-Based Exceptions
    """

class Sender:

    _discount=0
    _maxDiscount=0.25

    def __init__(self,senderID,name,address):
        self._senderID = senderID.upper()
        self._name = name
        self._address = address

    @property
    def senderID(self):
        return self._senderID
    @classmethod
    def discount(cls):
        return cls._discount
    @classmethod
    def setDiscount(cls,newValue):  #This method modifies the discount offered to senders
        #if new amount is greater than the max discount limit or is negative exception is raised
        #else new amount is the new discount amount
        if newValue > cls._maxDiscount:
            raise DeliveryException(f'Discount given cannot be greater than {cls._maxDiscount}')
        elif newValue<0:
            raise DeliveryException('Discount given cannot be negative')
        else:
            cls._discount=newValue
    def __str__(self):
        return f"Sender Id: {self._senderID} Name: {self._name} Address: {self._address}"

class RetailSender(Sender):

    _discount = 0.15
    _maxDiscount = 0.5

    def __init__(self,senderID,name,address,business):
        super().__init__(senderID,name,address)
        self._business = business
    def __str__(self):
        return f"Retail Business {self._business} {super().__str__()}"

from abc import ABC , abstractmethod ,abstractclassmethod

class ItemDelivery(ABC): #Abstract Class

    _nextID = 1

    def __init__(self,sender,weight,pickupDate,destination,express):
        self._deliveryID = ItemDelivery._nextID
        self._sender = sender
        #If weight to be send is negative or greater than the weight limit
        #exception is raised accordingly
        if weight<0:
            raise DeliveryException(f'Item weight {weight}kg cannot be negative')
        elif weight > type(self).getMaximumWeight():
            raise DeliveryException(f'Item weight {weight}kg exceed maximum weight {type(self).getMaximumWeight()}kg')
        else:
            self._weight = weight
        self._pickupDate = pickupDate
        self._destination = destination
        self._express = express
        ItemDelivery._nextID+=1  #keep on increasing delivery id number each time there is new item to deliver
        self._deliveredDate = None

    @property
    def sender(self):
        return self._sender
    @property
    def weight(self):
        return self._weight
    @property
    def pickupDate(self):
        return self._pickupDate
    @property
    def destination(self):
        return self._destination
    @property
    def express(self):
        return self._express
    @property
    def deliveryID(self):
        return self._deliveryID
    @property
    def deliveredDate(self):
        return self._deliveredDate
    @deliveredDate.setter
    def deliveredDate(self,newValue):
        self._deliveredDate=newValue

    @abstractclassmethod
    def getMaximumWeight(cls):
        pass
    @abstractmethod
    def expectedDeliveredDate(self):
        pass
    @abstractmethod
    def nonExpressCharges(self):
        pass
    @abstractmethod
    def expressCharges(self):
        pass

    def __str__(self):

        charges=None
        #Converting pickup date from datetime object into a string
        pickup=datetime.strftime(self._pickupDate,"%a, %d %b %Y")
        #Converting expected delivery date from datetime object into a string
        delivereddate=datetime.strftime(self.expectedDeliveredDate(),"%a, %d %b %Y")
        c_status=None   #variable which gets updated depending on delivery status of item
        if self._deliveredDate!=None:
            c_status = datetime.strftime(self._deliveredDate,"%a, %d %b %Y")
        else:
            c_status = 'Not Collected, in transit'
        #checks whether delivery is on express mode or not and makes charges based on that
        if self._express:
            charges = self.expressCharges() *  (1-type(self._sender).discount())
        else:
            charges = self.nonExpressCharges() * (1-type(self._sender).discount())
        yesorno=None
        if self._express:
            yesorno='Yes'
        else:
            yesorno="No"

        return f"Delivery ID: {self._deliveryID}"\
    f"\nExpress: {yesorno} Weight: {self._weight}kg Charges: ${charges:0.2f}"\
    f"\n        Pick Up: {pickup} Expected delivery date: {delivereddate}"\
    f"\n        Collected: {c_status} Destination: {self._destination}"\
    f"\n        {self._sender.__str__()}"


class LocalItemDelivery(ItemDelivery): #Class for senders who want to send items locally

    _turnAround = 1
    _expressRate = 1.25
    _base = 3

    @classmethod
    def getMaximumWeight(cls):
        return 40

    def expectedDeliveredDate(self):   #Calculates expected Delivery Date of item
        if self.express:
            return self.pickupDate + timedelta(days=(type(self)._turnAround/2))
        else:
            return self.pickupDate + timedelta(days=(type(self)._turnAround))

    def nonExpressCharges(self):   #Charges Calculation if Delivery is not made on express mode
        if self.weight < 3 :
            return type(self)._base + 3
        elif 3 <= self.weight < 5:
            return type(self)._base + 7
        elif 5 <= self.weight < 10 :
            return type(self)._base + 12
        else:
            return type(self)._base + 30

    def expressCharges(self):     #Charges Calculation if Delivery is made on express mode
        return self._expressRate * self.nonExpressCharges()

    def __str__(self):
        return f"Local Item Delivery {super().__str__()}"


class OverseasItemDelivery(ItemDelivery):    #Class for senders who want to send items Internationally

    _expressRate = 1.4
    _base = 10

    def __init__(self,sender,weight,pickupDate,destination,express,zone):
        #Checks For the Zone , if its within the range value will be assigned to the variable
        #else raises an exception
        if zone<=0 or zone>6:
            raise DeliveryException("Zone must be between 1 and 6")
        else:
            self._zone = zone
        super().__init__(sender,weight,pickupDate,destination,express)


    @classmethod
    def getMaximumWeight(cls):
        return 30

    def expectedDeliveredDate(self):    #Calculates expected Delivery Date of item
        if self.express:
            return self.pickupDate + timedelta(days = (2*self._zone))
        else:
            return self.pickupDate + timedelta(days = (3*self._zone))

    def nonExpressCharges(self):       #Charges Calculation if Delivery is not made on express mode
        if self.weight < 3 :
            return 30 + type(self)._base * self._zone/10
        elif 3 <= self.weight < 5:
            return 50 + type(self)._base * self._zone/10
        elif 5 <= self.weight < 10 :
            return 90 + type(self)._base * self._zone/10
        else:
            return 150 + type(self)._base * self._zone/10

    def expressCharges(self):        #Charges Calculation if Delivery is made on express mode
        return (self._expressRate + self._zone/10) * self.nonExpressCharges()
    def __str__(self):
        return f"Overseas Item Delivery Zone: {self._zone} {super().__str__()}"


class DeliveryCompany:

    def __init__(self):

        self._senders = {}     #Dictionary to store senders
        self._itemDelivery = []    #List To store items

    def searchSender(self,senderid):
        """
        Method Searches & returns registered sender in the Company if search is successful
        else returns None
        """
        if senderid in self._senders:
            return self._senders[senderid]
        else:
            return None

    def addSender(self,sender):
        """
        Method add Unique Sender in the Company
        if the addition is unsuccessful exception is raised
        """
        if self.searchSender(sender.senderID)==None:
            self._senders[sender.senderID]=sender
        else:
            raise DeliveryException(f"Sender {sender.senderID} is already recorded")

    def addDeliveryItem(self,itemDelivery):
        """
        Method add items to be delivered in the Company
        """
        self._itemDelivery.append(itemDelivery)

    def deliverItem(self,deliveryID):
        """
        Method returns True if Item is successfully delivered
        else an exception is raised
        """
        for item in self._itemDelivery:
            if item.deliveryID == deliveryID:
                item.deliveredDate= datetime.today()
                return True
        else:
            raise DeliveryException(f"Invalid delivery Id {deliveryID}")

    def itemDeliveriesStr(self):
        """
        Method returns String of item deliveries
        """
        s=""
        for item in self._itemDelivery:
            if "Overseas" in item.__str__():
                continue
            else:
                s+=item.__str__() + "\n"
        for item in self._itemDelivery:
            if "Overseas" in item.__str__():
                s+=item.__str__() + "\n"
            else:
                continue
        return f"Delivery List:\n{s}"
    def sendersStr(self):
        """
        Method returns String of senders
        """
        s=""
        for sender in self._senders:
            if 'Retail' in self._senders[sender].__str__():
                s+=self._senders[sender].__str__() + '\n'
            else:
                continue
        for sender in self._senders:
            if 'Retail' in self._senders[sender].__str__():
                continue
            else:
                s+=self._senders[sender].__str__() + '\n'

        return f"Senders:\n{s}"
    def __str__(self):
        return f"{self.sendersStr()}\n{self.itemDeliveriesStr()}"

def menu():
    """
    Function prints Menu for the Company
    """
    print("""Menu
1. List Senders
2. List Deliveries
3. Add Sender
4. Request Delivery Pickup
5. Deliver Item
0. Exit""")

def deliveritem(co):
    """
    Function Executes deliverItem method of Company Class
    """
    while True:
        try:
            s_id = int(input('Enter delivery id: ').strip()) #Asks for VALID Delivery ID
            co.deliverItem(s_id)
            break
        except ValueError:
            print("Please Enter a number")
        except DeliveryException as e:
            print(e)
    print(co.sendersStr())
    print('')
    print(co.itemDeliveriesStr())

def addSender(co):
    """
    Function Exeecutes addSender method of Company Class
    """
    while True:
        while True:
            s_id=input('Enter sender identification: ').upper().strip() #Ask for Sender's ID
            if s_id=='':
                continue
            else:
                break
        s_name=input('Enter sender name: ').strip()   #Ask for Sender's Name
        s_address=input('Enter address: ').strip()    #Ask for Sender's Address
        s_type=input('Enter type of sender - 1 for individual and 2 for retail: ').strip()  #Ask Sender's Type
        try:
            if s_type=='1':    #If sender's type is 1 we TRY to add him as a normal Sender
                sender=Sender(s_id,s_name,s_address)
                co.addSender(sender)
                print(f"{sender} added!")
                break
            elif s_type=='2':   #If sender's type is 2 we TRY to add him as a Retail Sender
                s_retail_co=input('Enter type of goods being retailed: ')
                sender=RetailSender(s_id,s_name,s_address,s_retail_co)
                co.addSender(sender)
                print(f"{sender} added!")
                break
        except DeliveryException as e:
            print(e)
            print('You have two attempts to correct sender identification')
            tries=1
            while tries!=3:  #This ensures only two attempts are provided to input Unique Sender's ID
                s_id = input(f'Attempt {tries} - Try another sender identification or <ENTER> to abandon: ').upper().strip()
                if s_id=='':    #If <Enter> is pressed we exit and gets redirected to the MENU page
                    print('Sender cannot added due to duplicate identification')
                    break
                else:
                    try:
                        if s_type=='1':
                            sender=Sender(s_id,s_name,s_address)
                            co.addSender(sender)
                            print(f"{sender} added!")
                            break
                        elif s_type=='2':
                            sender=RetailSender(s_id,s_name,s_address,s_retail_co)
                            co.addSender(sender)
                            print(f"{sender} added!")
                            break
                    except DeliveryException as e:
                        print(e)
                        if tries==2:
                            print("Cannot add sender")
                        tries+=1

            break


def requestDeliveryPickup(co):
    """
    Function Executes addDeliveryItem method of Company Class
    """
    print('Please enter correct sender id within 3 attempts')
    tries=1
    while tries!=4:  #Only have 3 attempts to input VALID Sender's ID once over user get directed to Menu Page
        s_id=input(f'{tries}. Enter sender id <ENTER> to end: ').upper().strip()
        if co.searchSender(s_id) == None:
            print('Invalid Sender ID')
            if tries==3:
                print('Please register the sender first')
            tries+=1
        else:
            while True:
                try:
                    s_w=float(input('Enter weight of item to deliver: ').strip()) #Ask for item's Weight
                    if s_w<=0:
                        raise Exception("Please enter a positive number for weight")
                    else:
                        break
                except ValueError:
                    print('Please enter a number')
                except Exception as e:
                    print(e)
            while True:
                try:
                    s_d=input('Enter pick up date in d/m/YYYY format: ').strip() #Asks For pickup Date
                    s_d=datetime.strptime(s_d,'%d/%m/%Y')
                    break
                except Exception:
                    print('Please enter pick up date in d/m/YYYY format')

            s_a=input('Enter destination address: ').strip()  #Asks for Sender's Address

            while True:
                s_e=input('By express? y for yes: ').strip()  #Asks whether item is to be delivered on Express Mode or Not
                if s_e=='':
                    continue
                else:
                    if s_e[0].lower()=='y':
                        s_e = True
                        break
                    else:
                        s_e = False
                        break
            while True:
                #Asks for the type of Delivery (Local/Overseas)
                s_over_local = input('Enter type of delivery - 1 for local and 2 for overseas: ').strip()
                if s_over_local!='1' and s_over_local!='2':
                    print('Please enter either 1 for local or 2 for overseas')
                else:
                    break
            if s_over_local=='1':
                try:
                    #TRY addiding the item into self._itemDelivery list in Class
                    d=LocalItemDelivery(co.searchSender(s_id),s_w,s_d,s_a,s_e)
                    co.addDeliveryItem(d)
                    print(d)
                    break
                except DeliveryException as e:
                    print(e)
                    while True:
                        try:
                            s_w=float(input('Enter weight of item to deliver: ').strip())
                            if s_w==0:
                                print("Please enter a positive number for weight")
                            else:
                                d=LocalItemDelivery(co.searchSender(s_id),s_w,s_d,s_a,s_e)
                                co.addDeliveryItem(d)
                                print(d)
                                break
                        except ValueError:
                            print('Please enter a number')
                        except DeliveryException as e :
                            print(e)
                    break
            elif s_over_local=='2':
                while True:
                    try:
                        #Asks for the Zone number
                        s_zone = int(input('Enter zone of destination: ').strip())
                        if s_zone<=0 or s_zone>6:
                            print('Zone must be between 1 and 6')
                            continue
                        else:
                            #TRY addiding the item into self._itemDelivery list in Class
                            d=OverseasItemDelivery(co.searchSender(s_id),s_w,s_d,s_a,s_e,s_zone)
                            co.addDeliveryItem(d)
                            print(d)
                            break
                    except ValueError:
                        print("Please enter a number")
                    except DeliveryException as e:
                        print(e)
                        while True:
                            try:
                                s_w=float(input('Enter weight of item to deliver: ').strip())
                                if s_w==0:
                                    print("Please enter a positive number for weight")
                                else:
                                    d=OverseasItemDelivery(co.searchSender(s_id),s_w,s_d,s_a,s_e,s_zone)
                                    co.addDeliveryItem(d)
                                    print(d)
                                    break
                            except ValueError:
                                print('Please enter a number')
                            except DeliveryException as e:
                                print(e)
                        break

                break

def main():


    #----As required by qns to create Company Class & populate it-----
    co=DeliveryCompany()
    s123 = Sender('S123','Peter Tan','12 Tampines')
    t111 = RetailSender('T111','Tan and Family','15 Tampines','Garment')
    co.addSender(s123)
    co.addSender(t111)
    co.addDeliveryItem(LocalItemDelivery(s123,4,datetime(2019,10,13),'1 Delivery Road',True))
    co.addDeliveryItem(LocalItemDelivery(s123,4,datetime(2019,10,13),'1 Delivery Road',False))
    co.addDeliveryItem(OverseasItemDelivery(s123,4,datetime(2019,10,13),'1 Some Land',True,1))
    co.addDeliveryItem(OverseasItemDelivery(s123,4,datetime(2019,10,13),'1 Some Land',False,1))
    co.addDeliveryItem(LocalItemDelivery(t111,4,datetime(2019,10,13),'1 Delivery Road',True))
    co.addDeliveryItem(LocalItemDelivery(t111,4,datetime(2019,10,13),'1 Delivery Road',False))
    co.addDeliveryItem(OverseasItemDelivery(t111,4,datetime(2019,10,13),'1 Some Land',True,1))
    co.addDeliveryItem(OverseasItemDelivery(t111,4,datetime(2019,10,13),'1 Some Land',False,1))
    co.addDeliveryItem(OverseasItemDelivery(t111,4,datetime(2019,10,13),'1 Some Land',False,2))
    co.addDeliveryItem(OverseasItemDelivery(t111,4,datetime(2019,10,13),'2 Some Land',True,2))




    while True:
        menu()   #Display the Menu
        try:
            option=int(input("Enter Option: ").strip()) #Asks for the option to be selected
        except ValueError:
            print("Please enter a whole number")
            continue
        else:
            if option==0:       #If option selected is 0 application shutsdown
                print("Application Closed")
                break
            elif option==1:     #List all the Senders in the Company
                print(co.sendersStr())
            elif option==2:     #List all the Deliveries in the Company
                print(co.itemDeliveriesStr())
            elif option==3:     #Add Sender in the Company if successful
                addSender(co)
            elif option==4:     #Request for Delivery Pickup if successful
                requestDeliveryPickup(co)
            elif option==5:     #Delivers the Item if successful
                deliveritem(co)

main()
```

Write a GUI delivery calculator that compute charges based on volumetric weight or

```python
from tkinter import *
from tkinter.scrolledtext import ScrolledText

class DeliveryChargesCalculator(Frame):

    def __init__(self):

        #Frame
        Frame.__init__(self)
        self.master.title('Delivery Charges Calculator')
        self.master.geometry('480x275')

        #--------Creation of COMPONENETS---------

        #Labels
        self._length=Label(self,text="Length:")       #Length Label
        self._width=Label(self,text="Width:")         #Width Label
        self._height=Label(self,text="Height:")       #Height Label
        self._weight=Label(self,text="Weight (kg):")  #Weight Label
        #Entries
        self._el=Entry(self,width=20)  #Length Entry
        self._ew=Entry(self,width=20)  #Width Entry
        self._eh=Entry(self,width=20)  #Height Entry
        self._ewt=Entry(self,width=20) #Weight Entry
        #RadioButtons
        self._radv=StringVar() #variable for radiobutton
        self._radv.set('cm')
        self._cm=Radiobutton(self,text='cm',variable=self._radv,value='cm')
        self._inch=Radiobutton(self,text='inch',variable=self._radv,value='inch')
        #Buttons
        self._charge=Button(self,text='Calculate Charge',command=self.Charge) #Charge Button
        self._clear=Button(self,text='Clear',width=8,command=self.Clear)      #Clear Button
        #ScrollText
        self._scrolltxt = ScrolledText(self,height=7,width=57)
        self._scrolltxt.configure(state=DISABLED)

        #---------PLACING OF COMPONENETS----------

        #Placing of LABELS
        self._length.grid(row=0,column=0,padx=(90,0))
        self._width.grid(row=1,column=0,padx=(85,0))
        self._height.grid(row=2,column=0,padx=(90,0))
        #Placing of Entries
        self._el.grid(row=0,column=1,sticky=W)
        self._ew.grid(row=1,column=1,sticky=W)
        self._eh.grid(row=2,column=1,sticky=W)
        self._weight.grid(row=4,column=0,padx=(115,0))
        self._ewt.grid(row=4,column=1,sticky=W)
        #Placing of Radio Buttons
        self._cm.grid(row=3,column=1,pady=5,sticky=W)
        self._inch.grid(row=3,column=1,sticky=E,padx=(0,50))
        #PLacing of Buttons
        self._charge.grid(row=5,column=1,columnspan=1,sticky=E+W,pady=5,padx=(0,5))
        self._clear.grid(row=5,column=2,columnspan=2)
        #Placing of Scroll Text
        self._scrolltxt.grid(row=6,column=0,columnspan=5)

        self.grid()


    def CalculateWeight(self):
        """
        Calls this function in 'Charge' function.

        """

        #This function calculates volumetric weight based on which unit of dimension is chosen either 'cm' or 'inches'
        lengthe= float(self._el.get())
        widthe = float(self._ew.get())
        heighte= float(self._eh.get())
        if self._radv.get()=='cm':
            v_w=(lengthe*widthe*heighte)/(6000)
            return v_w
        elif self._radv.get()=='inch':
            v_w=(lengthe*widthe*heighte)/(366)
            return v_w

    def Charge(self):
        """
        Calls this function when 'CalcualteCharge' button is clicked.

        """
        self._scrolltxt.configure(state=NORMAL)
        charged_weight=None
        #this ccompares volumetric weight,calculated in CalculateWeight function, to the actual weight
        #& base charges on the higher weight

        if self.CalculateWeight()>=float(self._ewt.get()):
            charged_weight=self.CalculateWeight()
        else:
            charged_weight=float(self._ewt.get())

        #checks for the conditions as implemented in the question
        delivery_charge=None
        if charged_weight<3:
            delivery_charge=6
        elif 3<=charged_weight<5:
            delivery_charge=10
        elif 5<=charged_weight<10:
            delivery_charge=15
        else:
            delivery_charge=33
        self._scrolltxt.insert('end',"Next Parcel\n")
        self._scrolltxt.insert('end',f"{float(self._el.get()):0.1f}{self._radv.get()} x {float(self._ew.get()):0.1f}{self._radv.get()} x {float(self._eh.get()):0.1f}{self._radv.get()} = {float(self._el.get())*float(self._ew.get())*float(self._eh.get()):0.2f}{self._radv.get()}^3"\
                   f" or {self.CalculateWeight():0.2f}kg\n")
        self._scrolltxt.insert('end',f"Charges based on the higher of ({self.CalculateWeight():0.2f} , {float(self._ewt.get()):0.2f}) = {charged_weight:0.2f}kg\n")
        self._scrolltxt.insert('end',f"Delivery Charge = ${delivery_charge:0.2f}\n")
        self._scrolltxt.insert('end','\n')
        self._scrolltxt.configure(state=DISABLED)

    def Clear(self):
        """
        Calls this function when 'Clear' button is clicked

        """
        #Remove everything from the enteries & Scrolledtext
        self._scrolltxt.configure(state=NORMAL)
        lis=[self._el,self._ew,self._eh,self._ewt]
        for x in lis:
            x.delete(0,'end')
        self._scrolltxt.delete(1.0,END)
        self._scrolltxt.configure(state=DISABLED)

def main():

    DeliveryCalulator = DeliveryChargesCalculator()
    DeliveryCalulator.mainloop()

main()
```
