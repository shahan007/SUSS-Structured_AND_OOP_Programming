```python
def tokenise(expression):

    """
    Tokenises the expression and return its tokens in a list
    """

    #Expression entered by the user is tokenised into a list by the removal of whitespace
    #This function will return tokens of the tokenised expression in a list

    token_list=expression.split(" ")
    return token_list

def check_variable(token_list):

    """
    Checks for variable in the expression.
    Returns False if there are variables in the expression it will return else Return True
    """
    

    v=[]
    for operand in token_list :
        if operand.isalpha():
            #If there is an alphabet in the expression it will append it to a list named v
            v.append(operand)
            #This will replace the variable with word 'and' in the expression
            token_list[token_list.index(operand)]=" and "
        else:
            continue

    #If the length of v is more than 1 or equals to 1 this will mean that expression has variable present
    if len(v)>=1:
        print(f"Invalid expression, expecting operand between {' '.join(token_list)}  but {','.join(v)} is a variable")
        #This is to empty the list and reset it so that this function can be used again
        v.clear()
        return False
    else:
        return True


def check_para(token_list):

    """
    Checks for unbalanced parentheses in the expression.
    Returns False if unbalanced .

    """

    #open_para list will collect ( , also known as left parentheses,that are present in the token list
    #close_para list will collect ) , also known as right parentheses ,that are present in the token list
    open_para=[]
    close_para=[]

    for para in token_list:
        if para=='(':
            open_para.append(para)
        elif para==")":
            close_para.append(para)
        else:
            continue

    if  token_list[0]=="(" and token_list[1]==")":
        print("Invalid expression, expecting operand between ( and )")
        return False
    #if the the first character in the expression is ) then we will excute the following message
    elif token_list[0][0]==")":
        print(f"Unexpected right parenthesis between and  {' '.join(token_list)}")
        return False
    #if there are more open parantheses,(, than we will excute the following message
    elif len(open_para)>len(close_para):
        print(f"There is {len(open_para)-len(close_para)} unclosed left parenthesis in {' '.join(token_list)}")
        return False
    #if there are more close parantheses,), than we will excute the following message
    elif len(close_para)>len(open_para):
        diff=len(close_para)-len(open_para)
        print(f"Unexpected right parenthesis between {' '.join(token_list[:-diff])} and {diff*')'} ")
        return False

    #if token in the expression is ) and the next token is ( we will ignore
    #as this will be performed in check_operand function
    for x in range(len(token_list)-1):
              if ")" in token_list[x] and "(" in token_list[x+1]:
                  break

def check_operand(token_list,expression):

    """
    Checks for missing operand or operator in the expression.
    Returns False if there are missing operand or operator in the expression.

    """
    #if the expression entered is equal to ) ( or ( ) or if the first token starts with )
    #we ignore it since its being handled in check_para , we dont want to repeat the message displayed

    if expression==") (" or expression=='( )' or expression[0][0]==')':
        pass
    else:
        for x in range(len(token_list)):
            #If element in list containing token is "" then we break out of this for loop
            if token_list[x]=="":
                break
            #Last element of the token list can be ignored
            if x==len(token_list)-1:
                pass
            else:
                #if the last  chracter of the token is ) and the next token has first character which is (
                #then we add in "and" between the two tokens
                if token_list[x][-1]==")" and token_list[x+1][0]=="(":
                    token_list.insert(x+1,'and')
            #if the token is ( and the next token contains ) then we will add in 's' between two tokens
            if token_list[x]=="(" and ")" in token_list[x+1]:
                token_list.insert(x+1,'s')
            #if the first chracter of the token is ( and the following chracter of the same token
            #is a numerical value then we add 's' between this and next token
            if token_list[x][0] == "(" and token_list[x][1:2] in ['-' ,'1','2','3','4','5','6','7','8','9'] :
                token_list.insert(x+1,'s')



    #If there is 'and' in the joined expression it means its missing an operator so return False
    if 'and' in " ".join(token_list):
        print(f'Invalid expression, expecting operator between {" ".join(token_list).replace("s","")}')
        return False
    #If there is 's' in the joined expression it means its missing an operand return False
    elif 's' in " ".join(token_list):
        print(f'Invalid expression, expecting operand between {" ".join(token_list).replace("s","and")}')
        return False

def check_operator(exp):

    """
    Checks for invalid usage of operator in the expression.
    Returns False if there is invalid usage of operator
    """

    for x in range(len(exp)-1):

        #if consecutive tokens in the token list are operators then the ones that are invalid
        #will result in an error message, whereas ones that are considered valid by the question
        #will be ignored
        if exp[x] in ['+','//','-'] and exp[x+1] in ['+','*','/','//','-']:
            exp.insert(x+1,'and')
        elif exp[x]=='/' and exp[x+1] in ['+','*','//','-']:
            exp.insert(x+1,'and')
        elif exp[x]=='*' and exp[x+1] in ['+','/','//','-']:
            exp.insert(x+1,'and')
        elif exp[x]=='*' and exp[x+1]=='*':
            pass
        elif exp[x]=='/' and exp[x+1]=='/':
            pass
        else:
            continue


    if 'and' in " ".join(exp):
        print(f'Invalid expression, expecting operand between {" ".join(exp)}')
        return False


def main():

    """
    Function prompts the user to input a mathematical expression in which each character is separated by a white space.
    It evaluates a mathematical expression given that it is valid.

    However if the mathematical expression entered is INVALID, a message will be displayed addressing the ERROR committed.

    Press <Enter> to quit the function

    """

    while True :
        expression=input("Enter Expression or <Enter> to quit: ")
        #if input entered is <enter> we quit the program else we will execute the expression
        if expression=="":
            print("Application Ended")
            break
        token_list=tokenise(expression)

        #here we will check for presence of variable , if its there we print the error and continue the loop
        #to ask the user to input again, else we will evaluate the expression
        if check_variable(token_list)==False:
            continue
        #here we look for presence of invalid consecutive operator ,if its there we print the error
        #and continue the loop to ask the user to input again, else we will evaluate the expression
        if check_operator(token_list)==False:
            continue
        #here we check for absence of a required operand and operator required to validate the expression.
        #if its there we print the error and continue the loop
        # to ask the user to input again, else we will evaluate the expression
        if check_operand(token_list,expression)==False:
            continue
        #here we check for unbalanced parantheses
        #if its there we print the error and continue the loop
        # to ask the user to input again, else we will evaluate the expression
        if check_para(token_list)==False:
            continue

        else:
            print(f"The answer is {eval(''.join(token_list)):0.1f}")
main()
```

```python
from random import shuffle, randint

def players(no_0f_players):
    
    """
    Function registers players name.
    Minimum Three players to be registered and maximum of Five players.
    Players Name needs to be Unique.

    """
    #Use while loop to continously get at least three players
    #and up till five players,this ensrues the user get prompted
    #until conditions are met.Names are being added into no_of_players Dictionary
    while True:

        player_1name=input("Enter player name, <Enter> to end : ")
        #If player name is <enter> we will put them into a while loop
        #of reasking them to key in a name
        while player_1name=="":
            print("Minimum 3 players please, currently only 0 player!")
            player_1name=input("Enter player name, <Enter> to end : ")

        no_0f_players['player_1']=player_1name
        #If player name is <enter> we will put them into a while loop
        #of reasking them to key in a name
        player_2name=input("Enter player name, <Enter> to end : ")
        while player_2name=="" or player_2name==player_1name:
            print("Minimum 3 players please, currently only 0 player!")
            player_2name=input("Enter player name, <Enter> to end : ")
        no_0f_players['player_2']=player_2name
        #If player name is <enter> we will put them into a while loop
        #of reasking them to key in a name
        player_3name=input("Enter player name, <Enter> to end : ")
        while player_3name=="" or player_3name==player_1name or player_3name==player_2name:
            print("Minimum 3 players please, currently only 0 player!")
            player_3name=input("Enter player name, <Enter> to end : ")
        no_0f_players['player_3']=player_3name
        #If there is no 4th player and <enter> is pressed we will exit the loop
        #else we will get into the loop
        while True:
            player_4name=input("Enter player name, <Enter> to end : ")
            if player_4name=="":
                break
            elif player_4name==player_1name or player_4name==player_2name or player_4name==player_3name:
                print("Not unique name")
                continue
            elif player_4name!=player_1name or player_4name!=player_2name or player_4name!=player_3name:
                no_0f_players['player_4']=player_4name
                break
        #If there is no Fourth player so  when <enter> is pressed then we will quit the loop
        if player_4name=='':
            break
        #If input is not an <enter> we will ask them to input Fifth players name
        #and put them into a while loop
        while True:
            player_5name=input("Enter player name, <Enter> to end : ")
            if player_5name=="":
                break
            elif player_5name==player_1name or player_5name==player_2name or player_5name==player_3name or player_5name==player_4name:
                print("Not Unique name")
                continue
            elif player_5name!=player_1name or player_5name!=player_2name or player_5name!=player_3name or player_5name!=player_4name:
                no_0f_players['player_5']=player_5name
                break

        break
    print(f"{len(no_0f_players)} players playing this game")


def player_points(player_point_details,no_0f_players):
    """
    Function initialise points for all players to 2

    """
    #We look in no_of_players Dictionary and from there look for players and
    #then create player_point_details dictionary which stores all players with
    #Point 2
    for x in no_0f_players:
        player_point_details[no_0f_players[x]]=2


def player_turns(no_0f_players):
    """
    Function returns list of players order in which the game will be played
    """
    #Get values from no_of_players Dictionary and convert into a list
    players_order=list(no_0f_players.values())    
    #This allows players name in the list made to randomly shuffle their order of game play
    shuffle(players_order)
    print("Playing in this order:",tuple(players_order))
    print("Start game. Roll 2 dice per player")
    return players_order

def dice_auto():
    """
    Function returns random value of First and Second Dice and their sum

    """    

    dice1=randint(1,6)
    dice2=randint(1,6)
    sum_=dice1+dice2
    return dice1,dice2,sum_

def dice_manual(dic_total_player,players_orders,dic):
    """
    Function asks for player to either roll or miss the dice.
    Enter r to roll the dice else enter m not to roll the dice and skip the turn

    If the total of all the dices played is more than 13, player will automatically
    have no more tries and be BUSTED!

    """    

    for x in players_orders:
        while True:
            roll_or_miss=input(f"{x} roll or miss ? (r/m)? ").lower()
            if roll_or_miss!="r" and roll_or_miss!="m":
                continue
            #In the main function there is going to be an empty dictionary named 'dic'
            #to store the dice values in a list for respective players
            dice=dic[x]
            total=sum(dice)
            #If roll_or_miss is 'm' we will skip the players turn and print his status
            if roll_or_miss=="m":
                print((f"{x} , Dice : {tuple(dice)} total  {total}"))
            while roll_or_miss[0]=="r":
                #If player choose to roll his value obtained is added into the list dice which has
                #previous dice values too
                dice.append(randint(1,6))
                #We will then calculate the players dice values total to determine
                #whether he gets to roll again or gets busted if his total exceeds 13
                total=sum(dice)
                #When total dices value of a player is less than 13
                #it means he has not been busted and still has the option to roll or miss
                if total<=13:
                    print((f"{x} , Dice : {tuple(dice)} total  {total}"))
                    while True:
                        roll_or_miss=input(f"{x} roll or miss ? (r/m)? ").lower()
                        if roll_or_miss=="r" or roll_or_miss=="m":
                            break
                        else:
                            continue
                    if roll_or_miss[0]=="m":
                        break
                    elif roll_or_miss[0]=='r':
                        continue
                #The moment total players dice values is more than 13 we break out of the while loop
                #and hand over the turn to the next player
                elif total>13:
                    print((f"{x} , Dice : {tuple(dice)} total  {total} Bust!"))
                    break
            #Players total dice values will be stored in dictionary named dic_total_player
            dic_total_player[x]=total
            break

def game_summary(dic,dic_total_player,player_point_details):
    """
    Function displays Game Summary of all the players progress
    If player wins he gets two point added to his score
    If player is Busted two point are then deducted from his score

    """
    #We create a dictionary to contain players whose total dice score is less than 13
    dic_total_less_13={}
    for x in dic_total_player:
        if  dic_total_player[x]<=13 :
            dic_total_less_13[x]=dic_total_player[x]
    #We look for player who has the highest Dice Total score from
    #dic_total_less_13 dictionary
    highest=max(dic_total_less_13,key=dic_total_less_13.get)
    for x in dic_total_player:
        if dic_total_player[x]<=13:
            #When the Player Dice Total equals highest value he is displayed as a winner
            #and two points are added to his Points and his values get updated in
            #player_point_details dictionary
            if dic_total_player[x]==dic_total_player[highest]:
                print(f"{x} , dice {tuple(dic[x])} , total {dic_total_player[x]}.  Winner ! Points : {player_point_details[x]} + 2 = {player_point_details[x]+2} ")
                player_point_details[x]+=2
            #When the Player Dice Total is not equal to highest value his values remain unchanged
            elif  x!=highest :
                print(f"{x} , dice {tuple(dic[x])} , total {dic_total_player[x]}. Points : {player_point_details[x]} ")
        #When the Player Dice Total is more than 13 we deduct two points from his points tally
        #and he is displayed as BUSTED!.His Values get updated in player_point_details dictionary
        elif dic_total_player[x]>=14:
            print(f"{x} , dice {tuple(dic[x])} , total {dic_total_player[x]}. BUST!  Points : {player_point_details[x]} - 2 = {player_point_details[x]-2}")
            player_point_details[x]-=2


def check_points(player_point_details):
    """
    Function checks for players points and also asks for the continuation of the game.
    Returns False if player total point is 0
    Returns False if player total point is 10

    """

    for x in player_point_details:
        #When players total points are 0 we quit the loop and return False
        if player_point_details[x]==0 :
            print("Player Bust ! End Game")
            return False
            break
        #When players total points are 10 we quit the loop and return False
        elif player_point_details[x]==10 :
            print("Reached limit of 10 Points  ! End Game")
            return False
            break

        else:
            continue
    #We ask if they want to play again
    ask=input("Continue next game (y/n) ? : ").lower()
    #if input is neither 'y' nor 'n' we keep on asking them until they enter
    #correct value which is 'y' or 'n'
    while ask[0]!='y' and ask[0]!='n':
        ask=input("Continue next game (y/n)? : ").lower()
    if ask=='y':
        return True
    else:
        print("End game")
        return False


    
def main():
    
    """
    Function plays Black Jack Game

    Rules of the game:
    - Minimum of 3 players and Maximum of 5 ( Unique names only )
    - If player chooses r he rolls else if he chooses m he skips his turn
    - If player Total Dice Score is less than 13 he can further choose to either roll or miss
    - If player Total Dice Score is more than 13 he is BUSTED and turn will be given to next player
    - If player is not Busted and is the HIGHEST he will be the winner and earn two points
    - If player is not Busted and is not the HIGHEST his values remains unchanged
    - If player is Busted two points will be deducted from his points tally
    - If player total point tally is either 0 or 10 , game will end

    """
    no_0f_players={}
    player_point_details={}
    dic_total_player={}
    
    #We get players name in
    players(no_0f_players)
    #We assigned players points to be 2
    player_points(player_point_details,no_0f_players)
    while True:
        #We get the order players will be playing in
        players_orders = player_turns(no_0f_players)
        #dic is to store players each dice values in a list
        dic={}
        for x in players_orders:
            dice1,dice2,sum_=dice_auto()
            dice=[dice1,dice2]
            dic[x]=dice
            print(f"{x} , Dice : {(dice1,dice2)} total  {sum_}")
        #Ask players to either roll or miss and act according to the function
        dice_manual(dic_total_player,players_orders,dic)
        print("GAME SUMMARY :")
        game_summary(dic,dic_total_player,player_point_details)
        #If check_points function returns False we quit the game else we continue
        if not check_points(player_point_details):
            break

main()
```
