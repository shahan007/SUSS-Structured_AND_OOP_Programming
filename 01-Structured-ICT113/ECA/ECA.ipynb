```python
def userName():
    """
    Function returns user name and their tier
    """
    name = input("Enter the name of the participant: ")
    tier = int(input("Enter tier that participant is currently in:"))

    #If tier value is more than 6 or less than zero
    #user is repeatedly asked for the tier value
    while tier>6 or tier<0:
        print('Invalid tier. Please enter 0 to 6')
        tier = int(input("Enter tier participant is currently in:"))
    return name,tier

def collectHealthPoint(name,tier):
    """
    Function returns health points
    """
    #If tier value is 6 return 0
    if tier == 6:
        print(f"{name}  is not eligible for any voucher as has made 6 redemption")
        return 0
    else:
        #Ask for users health points
        health_points = int(input('Enter health points that participant currently has: '))
        #If health points is less than or equal to zero return 0
        if health_points<=0:
            print(f"{name}  is not eligible for any voucher as he has invalid {health_points} points")
            return 0
        return health_points

def calculateRedemptions(points,name,tier,redemption_amount):
    """
    Function calculates redemption amount and returns bool
    """

    #If tier is equal to zero
    if tier[0]==0:
        if points[0]>=750:  #If user points are more than or equal to 750
            old_points = points[0]  #User old points
            points[0]-=750  # substract points used to redeem a voucher
            tier[0]+=1     # add up the tier value by one
            redemption_amount[0]+=5   #amount redeemed increase by 5
            print(f"Participant {name} gets $5 voucher. Current Health point: {old_points}. Deducted Health points: 750. Final Health points:{points[0]}")
            return True
        else:
            #If points become 0
            if points[0]==0:
                print(f'Participant {name} has no health points')
            else:
                #print Insufficient amount message
                print(f"{points[0]} points is insufficient for tier {tier[0]+1} redemption. Require 750 health points")
            return False

    #If tier is equal to one
    elif tier[0]==1:
        if points[0]>=1500:         #If user points are more than or equal to 1500
            old_points = points[0]  #User old points
            points[0]-=1500         # substract points used to redeem a voucher
            tier[0]+=1              # add up the tier value by one
            redemption_amount[0]+=10  #amount redeemed increase by 10
            print(f"Participant {name} gets $10 voucher. Current Health point: {old_points}. Deducted Health points: 1500. Final Health points:{points[0]}")
            return True
        else:
            #If points become 0
            if points[0]==0:
                print(f'Participant {name} has no health points')
            else:
                #print Insufficient amount message
                print(f"{points[0]} points is insufficient for tier {tier[0]+1} redemption. Require 1500 health points")
            return False

    #If tier is equal to 2/3/4/5
    elif tier[0]==2 or tier[0]==4 or tier[0]==5 or tier[0]==3:
        if points[0]>=750:        #If user points are more than or equal to 750
            old_points=points[0]  #User old points
            points[0]-=750        # substract points used to redeem a voucher
            tier[0]+=1            # add up the tier value by one
            redemption_amount[0]+=5   #amount redeemed increase by 5
            print(f"Participant {name} gets $5 voucher. Current Health point: {old_points}. Deducted Health points: 750. Final Health points:{points[0]}")
            return True
        else:
            #If points become 0
            if points[0]==0:
                print(f'Participant {name} has no health points')
            else:
                #print Insufficient amount message
                print(f"{points[0]} points is insufficient for tier {tier[0]+1} redemption. Require 750 health points")
            return False
    elif tier[0]==6:
        #If tier value becomes 6 and health point by then is 0 then this will be executed
        if points[0]==0:
            print(f'Participant {name} has no health points')

def main():

    """
    Function redeems participants health points for a voucher and updates the new health points status
    """
    #Get name and tier
    name,tier = userName()

    #get healthpoint value
    healthpoint = collectHealthPoint(name,tier)

    #Checks If tier is 6 or healthpoints entered are ineligible
    #If False we cannot redeem
    if not healthpoint:
        pass
    else:
        redemption_amount = [0]    #list for amount to be redeem so it can be mutated
        tiero = [tier]             #list for tier so it can be mutated
        healthpoint = [healthpoint]   #list for healthpoints so it can be mutated


        while tiero[0]<7:
            #If cannot be redeem any further user exit else we continue the redemption process
            if not calculateRedemptions(healthpoint,name,tiero,redemption_amount):
                print(f"Total Redemption Amount = ${redemption_amount[0]}")
                break
            else:
                print(f"Participant {name} has moved to tier {tiero[0]} redemption")
                continue

        print(f"Participant {name} is in tier {tiero[0]} and has {healthpoint[0]} health points")

main()
```

```python
from random import choice

def createLevelQuestions(level,number_list):

    """
    Function sets the number list for the game depending on the game level.
    """

    #Depending on the level of the game the first number and the second number of the game
    #will be generated
    f_number = 10**(level-1) #firstNumber
    l_number = 10**level     #LastNumber

    #adds to the number list containg all the numbers within the range of the first number and last number
    number_list.extend(x for x in range(f_number,l_number))

def getOperator():

    """
    Function returns Operator for the game
    """

    #Operators list
    operator = ['+','-']

    #randomly returns an operator
    return choice(operator)

def getNumber(operator,number_list):

    """
    Function returns first number, second number
    """

    f_num = choice(number_list)   #Random first number
    s_num = choice(number_list)   #Random Second Number

    #If both numbers are not unique we will change the second number and randomly get unqiue one
    while s_num == f_num:
        s_num = choice(number_list)

    #If operator is '-' then first number must always be bigger than the second number
    if operator=='-':
        if s_num > f_num:
            return s_num,f_num                
    return f_num,s_num

def post_check(fNumber,sNumber,operator,tries,level_score):

    """
    Function checks players answer with the correct answer and print a review
    """

    #If operator is + we add the numbers
    if operator=='+':
        score = fNumber + sNumber
    else:
        #Else we substract the numbers
        score = fNumber - sNumber

    player_input = int(input(f"{tries}. {fNumber} {operator} {sNumber} = "))
    if player_input!=score:   #If player answer is wrong we print incorrect and give the right answer
        print(f"Incorrect, answer is {score}")
    else:
        level_score[0]+=1 #We increase the score by 1 for that level
        print('Correct!')

def replay(level):

    """
    Function checks whether player want to replay and returns bool
    """
    
    play = input(f'Press Enter to proceed to level {level+1} or Q to Quit:')

    #If play is not <Enter> or not Q/q  we ask for the value over again until right value is entered
    while play!='' and play.upper()!='Q':
        print("Invalid Entry!")
        play = input(f'Press Enter to proceed to level {level+1} or Q to Quit:')

    if not play: #If value is <Enter> return True else return False
        return True
    return False

def reviewLast(overallScore):

    """
    Function displays the score by each level at the end of the application
    """

    print('Scores:')

    for index,scores in enumerate(overallScore):
        print(f"Level {index+1}: {scores[0][0]}/{scores[1]}")

def main():

    #Initialise level to 1
    level = 1

    #Score List
    overallScore = []
    tries =  int(input('Enter number of questions per level:'))


    while True:

        number_list = []
        #Update Number List
        createLevelQuestions(level,number_list)
        #Number of questions the player want to play


        #Intialise
        t = 1

        level_score = [0] #Score Level

        print(f"Level {level}")

        while t<tries+1:

            operator = getOperator() #Get the operatot
            f_num,s_num = getNumber(operator,number_list) #Get the number
            post_check(f_num,s_num,operator,t,level_score) #Check and print the feedback
            t+=1
        print(f"Score: {level_score[0]}/{tries}")
        overallScore.append([level_score,tries]) #Append the level score to overall score
        if not replay(level): #If dont want to replay break out of the loop Else increase level by 1
            break
        level+=1
    #Display the final result of all levels
    reviewLast(overallScore)

main()
```

```python
def getNames():

    name_list = []  #name list

    while True:
        name = input("Enter player's name and <ENTER> to stop:").upper()

        #If name == ''
        if not name:
            if len(name_list)<2:  #If name list doesnt have two unique names yet this will continue back in while loop
                print('There must be at least two players')
                continue
            else:
                #Else if name list doesnt has two unique names this will break out of while loop
                break
        else:
            #if name not unqiue continue back in the while loop
            if name in name_list:
                print(f"{name} is already recorded. Please enter a different name.")
                continue
        name_list.append(name)  #append unique name into the name list
    return name_list

def createGameList(name_list):

    #Creates game list with player name from name list and intial score of 0
    game_list = []
    #add in name and 0 which is inital total score
    for name in name_list:
        game_list.append([[name,0]])
    return game_list

def appendGameList(game_list):
    #Game list with intial rounds of 4
    for players in game_list:
        players.extend([[0,0],[0,0],[0,0],[0,0]])

from random import randint

def throwDice(rounD,player_list):

    dice1= randint(1,6) #Randomly throw dice 1
    dice2= randint(1,6) #Randomly throw dice 2
    #specififc player list containg specific one players data
    #will have its dice value updated depending on the round
    player_list[rounD][0]=dice1
    player_list[rounD][1]=dice2

def calculateScore(player_score,player_list):
    #calculates sum of each rounds of a specific player and append it to players own player score list
    for scores in player_list[1:]:
        player_score.append(sum(scores))

from copy import deepcopy
def checkSimilarity(player_score,player_list):


    player_sc = deepcopy(player_score)  #creates deep copy of player score
    player_ls =  player_list[1:]    #since player_list[0] is player name hence we index from 1
    #Check if dice value for each throw under each round is same or not
    for x in range(len(player_ls)-1):
        #if same then player score for that round gets updated by adding the score of
        #subsequent round
        if player_ls[x][0]==player_ls[x][1]:
            player_sc[x]+=player_sc[x+1]

    player_score.clear() #clear the player_score list
    player_score.extend(player_sc)      #mutate it with player_sc created

def displayScore(player_list,player_score):

    print(f"Player {player_list[0][0]}")
    #If player_list has 5 rounds we execute the following by printing all 5 rounds
    if len(player_list)==6:
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}{5:<5}".format('Round','R1','R2','R3','R4','R5'))
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}{5:<5}".format('Throw1',player_list[1][0],player_list[2][0],player_list[3][0],player_list[4][0],player_list[5][0]))
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}{5:<5}".format('Throw2',player_list[1][1],player_list[2][1],player_list[3][1],player_list[4][1],player_list[5][1]))
    else:
        #If player_list has 4 rounds we execute the following by printing all 4 rounds
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}".format('Round','R1','R2','R3','R4'))
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}".format('Throw1',player_list[1][0],player_list[2][0],player_list[3][0],player_list[4][0]))
        print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}".format('Throw2',player_list[1][1],player_list[2][1],player_list[3][1],player_list[4][1]))

def Score(player_list):

    """
    Function prints the scores for each round
    """

    Score = player_list[-1]
    s=''
    for x in range(4):
        s+=f"{Score[x]:<5}"
    print("{0:<10}{1}".format('Score',s))

def accumulatedScore(accScoreList):
    """
    Function prints accumulated scores by the rounds.
    """

    f_= accScoreList[0]
    S_= accScoreList[1]+f_
    T_ = S_ + accScoreList[2]
    F_ = T_ + accScoreList[3]
    print("{0:<10}{1:<5}{2:<5}{3:<5}{4:<5}".format('Acc.sc',f_,S_,T_,F_))

def replay():

    """
    Function checks for replay and return bool
    """

    play= input('\nDo you want to play again? <ENTER> for no:')
    if play=='':
        return False
    return True

def winOrLoss(game_list):

    temp_list = deepcopy(game_list) #make deepcopy for game_list
    #Sort the temp_list by the total score and arrange in descending order
    temp_list.sort(key=lambda x : x[0][1],reverse = True)
    maximum = temp_list[0][0][1]  #maximum score
    #If players scores are equal to maximum their name will be appended to the win_list
    win_list = [x[0][0] for x in temp_list if x[0][1]==maximum]  #list for the winners
    strings = ' , '.join(win_list)
    print(f"Overall winner: {strings} with total score of {maximum}\n")

def main():

    #name list
    name = getNames()
    #create MAIN LIST containing game data
    game_list = createGameList(name)

    game = 1 #initialize game to 1


    while True:

        print(f"\nGame Summary for Game {game}")

        appendGameList(game_list) #Append the list with basic 4 rounds


        i=0 #index of game_list

        while i<len(game_list):

            player_list=game_list[i] #player list which is nested in the game_list

            initial_round=1
            rounds =5

            while initial_round<rounds:
                #Throw the dices and execute the function
                throwDice(initial_round,player_list)
                initial_round+=1
                if initial_round==5:
                    #if both dice value for round 4 are equal than we create round 5
                    if player_list[4][0]==player_list[4][1]:
                        player_list.append([0,0])
                        #Throw the dices fot round 5 as well
                        throwDice(5,player_list)
            player_score = [] #player_score list
            calculateScore(player_score,player_list) #calculates the score for each round and appends to the player_score
            print('\n')
            displayScore(player_list,player_score) #Display the data value
            checkSimilarity(player_score,player_list) #checks if dice values are equal and execute the function
            player_list.append(player_score) #append player_score list to player_list
            Score(player_list)  #prints the score value for the rounds
            if len(player_list)==7:
                player_list[-1]= player_list[-1][:-1] #Exclude the round 5 score value
            accScoreList = player_list[-1]  #Score list
            accumulatedScore(accScoreList)  #prints the accumulated score list
            print('\n')
            previous = player_list[0][1]  #previous total score
            player_list[0][1]+=sum(accScoreList)  #new total score
            print(f"Previous Total : {previous} Current Total : {player_list[0][1]}\n")
            game_list[i]= player_list
            game_list[i]= game_list[i][0:1] #get back the original state of game_list with just player name and total score
            i+=1 #increase the index by 1

        winOrLoss(game_list) #Display the winner of game
        if not replay():  #if dont want to continue playing exit the game
            break
        game+=1  #proceed to next game
    print('\n')
    winOrLoss(game_list) #Display the winner of game

main()
```
